```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import Imputer
```

```python
app_train=pd.read_csv("application_train.csv")
app_train.head()
pd.set_option('display.max_columns', None)
```

```python
app_train.shape
```

```python
app_train.describe()
```

```python
app_train['DAYS_EMPLOYED'].replace({365243: np.nan}, inplace = True)
```

```python
app_train.describe()
```

```python
app_train.iloc[:,1].dtypes
```

```python
for i in range(0,122):
    if(app_train.iloc[:,i].dtypes==object):
        print("col is : ",i,"  ",app_train.iloc[:,i].value_counts())
        print("\n\n")
```

```python
# getting the colunms in which we need to apply modeimpute. 
modeimpute=[]
for i in range(0,122):
    if(app_train.iloc[:,i].dtypes=='int64' or app_train.iloc[:,i].dtypes=='float64'):
        if(max(app_train.iloc[:,i])==1 and min(app_train.iloc[:,i])==0):
            if (x in [0,1] for x in app_train.iloc[:,i].value_counts()):
                modeimpute.append(i)
len(modeimpute)
```

```python
objectindex=[]
for i in range(0,122):
    if i not in modeimpute:
        objectindex.append(i)
print(len(objectindex))
```

```python
# newapptraindf=app_train
apptraindfintfloat01=app_train.drop(app_train.columns[objectindex],axis=1)
apptraindfintfloat01.shape
```

```python
apptraindfintfloat01.head()
# this will be mode imputed
```

```python
apptraindfintfloat01.isna().sum().sum()
```

```python
mean_imputer = Imputer(missing_values='NaN', strategy='most_frequent', axis=0)
mean_imputer = mean_imputer.fit(apptraindfintfloat01)
imputed_df = mean_imputer.transform(apptraindfintfloat01.values)
cols=apptraindfintfloat01.columns
app_train_11=pd.DataFrame(imputed_df,columns=cols)
app_train_11.head()
```

```python
app_train_11.isna().sum().sum()
```

```python
app_train_11.dtypes.value_counts()
```

```python
apptraindfobjectandcont=app_train.drop(app_train.columns[modeimpute],axis=1)
apptraindfobjectandcont.shape

pureobject=[]
for i in range(0,46):
    if(apptraindfobjectandcont.iloc[:,i].dtypes=="object"):
        pureobject.append(i)
print(len(pureobject))
```

```python
useful=[x for x in range(0,46) if x not in pureobject]
app_traindf_pureobj=apptraindfobjectandcont.drop(apptraindfobjectandcont.columns[useful],axis=1)
app_traindf_pureobj=pd.get_dummies(app_traindf_pureobj)
app_traindf_pureobj.head()

```

```python
app_traindf_pureobj.shape
```

```python
app_traindf_pureobj.isna().sum().sum()
```

```python
app_traindf_meanimpute=apptraindfobjectandcont.drop(apptraindfobjectandcont.columns[pureobject],axis=1)
app_traindf_meanimpute.describe()
print(app_traindf_meanimpute.isna().sum().sum())
```

```python
temptemp=app_traindf_meanimpute.iloc[:,[-11,-10,-9,-8,-6,-5,-4,-3,-2,-1]]
print(temptemp.isna().sum().sum())
mean_imputer = Imputer(missing_values='NaN', strategy='most_frequent', axis=0)
mean_imputer = mean_imputer.fit(temptemp)
imputed_df = mean_imputer.transform(temptemp.values)
cols=temptemp.columns
app_train_temptemp=pd.DataFrame(imputed_df,columns=cols)
# print(app_train_temptemp.head())
print(app_train_temptemp.isna().sum().sum())
```

```python
app_train_temptemp.shape
```

```python
app_traindf_meanimpute.isna().sum().sum()
```

```python
app_traindf_meanimpute.shape
```

```python
col1=[-11,-10,-9,-8,-6,-5,-4,-3,-2,-1]
col2=[-11+30,-10+30,-9+30,-8+30,-6+30,-5+30,-4+30,-3+30,-2+30,-1+30]
print(col2)
app_traindf_meanimpute2 = app_traindf_meanimpute.drop(app_traindf_meanimpute.columns[col2],axis=1)
print(app_traindf_meanimpute2.isna().sum().sum())
```

```python
app_traindf_meanimpute2.shape
```

```python
print(app_traindf_meanimpute2.isna().sum().sum())
mean_imputer = Imputer(missing_values='NaN', strategy='mean', axis=0)
mean_imputer = mean_imputer.fit(app_traindf_meanimpute2)
imputed_df = mean_imputer.transform(app_traindf_meanimpute2.values)
cols=app_traindf_meanimpute2.columns
app_train_14=pd.DataFrame(imputed_df,columns=cols)
app_train_14.head()
# print(app_train_14)
print(app_train_14.isna().sum().sum())
```

```python
app_train_14.shape
```

```python
app_train_14.isna().sum().sum()
```

```python
result = pd.concat([app_train_11,app_traindf_pureobj ,app_train_temptemp ,app_train_14], axis=1, sort=False)
result.head()
```

```python
result.shape
```

```python
result.to_csv("application_train_imputed.csv")
```
